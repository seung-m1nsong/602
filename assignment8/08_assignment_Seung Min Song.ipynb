```python
# 1. Modify multiple column names.
# change famsup to famEdsup, chage Dacl to dayAlc, change Walc to weekAlc, and change goout to friendtime
df_mat = df_mat.rename(columns={'famsup': 'famEdsup'})
df_mat = df_mat.rename(columns={'Dacl': 'dayAlc'})
df_mat = df_mat.rename(columns={'Walc': 'weekAlc'})
df_mat = df_mat.rename(columns={'goout': 'friendtime'})

df_por = df_por.rename(columns={'famsup': 'famEdsup'})
df_por = df_por.rename(columns={'Dacl': 'dayAlc'})
df_por = df_por.rename(columns={'Walc': 'weekAlc'})
df_por = df_por.rename(columns={'goout': 'friendtime'})

# 2. Look at the structure of your data – are any variables improperly coded? Such as strings or characters? Convert to correct structure if needed.
# The dataset has been cleaned and all columns have the appropriate data types.

# 3. Fix missing and invalid values in data.
# The dataset is complete with no missing values.


# 4. Create new columns based on existing columns or calculations.
# use the assign() method to create a new column named G1G2avg. Later I will compare this new column with G3 (Final) grade. 
df_mat = df_mat.assign(G1G2avg = (df_mat['G1'] + df_mat['G2'])/2)
print(df_mat.head())

df_por = df_por.assign(G1G2avg = (df_por['G1'] + df_por['G2'])/2)
print(df_mat.head())

# create a new column named 'class' with the value 'mat' and 'por'
df_mat['class'] = 'mat'
df_por['class'] = 'por'

#5. Drop column(s) from your dataset.
# drop df_mat index 1. However, I will store the data to df_new, because I don't want to modify the orginal data. 
#df_new = df_mat.drop(index=1)

#6. Drop a row(s) from your dataset.
# drop df_mat column 'school'. However, I will store the data to df_new, because I don't want to modify the orginal data. 
#df_new = df_mat.drop('school', axis=1)
#print(df_new.head())

#7. Sort your data based on multiple variables. 
# first I will combine the two dataset
df_student = pd.concat([df_mat, df_por], axis=0)
# Reset the index
df_student = df_student.reset_index(drop=True)
print(df_student)

# sort by sex first, and then by age using sort_values
#df_student.sort_values(['sex', 'age'], inplace=True)
#print(df_student)

#8. Filter your data based on some condition. 
# filter the data in df_student based on the condition where G3 grade is higher than G1G2avg
df_student_filtered = df_student[df_student['G3'] > df_student['G1G2avg']]
print(df_student_filtered)
# out of 1044 students, 561 students scored a higher grade than the combined average of G1 and G2.

#9. Convert all the string values to upper or lower cases in one column.
# use the str.lower() method to convert all the string values to lower case in the "school" column of df_student.
df_student["school"] = df_student["school"].str.lower()
print(df_student.head())

#10. Check whether numeric values are present in a given column of your dataframe.
#use the str.contains() method of pandas series along with a regular expression pattern to match numeric values.
#True means it contains numeric values.
numeric_present = df_student['famsize'].str.contains('\d').any()
print(numeric_present)

#11. Group your dataset by one column, and get the mean, min, and max values by group.

# • Groupby()
# • agg() or .apply() 
# group it with school and calculates the mean, min, and max values of the GIG2avg column for each group using the agg function. 
# school gp performed better than school ms.
df_group = df_student.groupby(['school']).agg({'G1G2avg': ['mean', 'min', 'max']})
print(df_group)

#12. Group your dataset by two columns and then sort the aggregated results within the groups.
# group it with school and class then calculates the mean, min, and max values of the GIG2avg column for each group using the agg function. 
df_group = df_student.groupby(['school', 'class']).agg({'G1G2avg': ['mean', 'min', 'max']})
# Sort the results within each group by descending order of the mean G1G2avg
df_group_sorted = df_group.sort_values(by=('G1G2avg', 'mean'), ascending=False)

print(df_group_sorted)

#13. Use the replace() method to replace values. Change all variable with Yes to 1  and No to 0.
df_student.replace({"yes": 1, "no": 0}, inplace=True)
print(df_student)

#14. Create a new DataFrame for the numeric columns and character columns separately.
df_n = df_student.select_dtypes(include='number')
df_c = df_student.select_dtypes(include='object')

df_student.to_csv('file_path/df_student.csv', index=False)
```

```python
# Part 1
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Graph 1 (boxplot)
# Group by sex and calculate mean age
df_agesex = df_student.groupby('sex')['age'].mean().reset_index()
print(df_agesex)
# The average age for females is 16.8 years and the average age for males is 16.7 years, based on the student dataset.

# group data by sex and find maximum age for each group
max_age = df_student.groupby('sex')['age'].max()

# print maximum age for each sex
print("Maximum age for Female: ", max_age['F'])
print("Maximum age for Male: ", max_age['M'])

# create figure and axis objects
fig, ax = plt.subplots()

# create boxplot
bp = ax.boxplot(x=[df_student[df_student['sex']=='F']['age'], df_student[df_student['sex']=='M']['age']], labels=['Female', 'Male'])

# change legend position and font size
ax.legend(loc='upper right', fontsize='small')

# place legend outside of plot
plt.subplots_adjust(right=0.8)
ax.legend(loc='center left', bbox_to_anchor=(1.05, 0.5))

# change title and x/y labels
ax.set_title('Age by Sex')
ax.set_xlabel('Sex')
ax.set_ylabel('Age')

# show plot
#plt.show()


# Graph 2(barplot)
df_c = df_student.select_dtypes(include='object')
print(df_c)


# Subset the dataframe to only include females taking Por or Mat
female_por_mat = df_c[(df_c['sex'] == 'F') & ((df_c['class'] == 'por') | (df_c['class'] == 'mat'))]

# Count the number of students in each class
counts_f= female_por_mat['class'].value_counts()

print(counts_f)

# Subset the dataframe to only include males taking Por or Mat
male_por_mat = df_c[(df_c['sex'] == 'M') & ((df_c['class'] == 'por') | (df_c['class'] == 'mat'))]

# Count the number of students in each class
counts_m= male_por_mat['class'].value_counts()

print(counts_m)

# Create a figure with two subplots
fig, axs = plt.subplots(1, 2, figsize=(10, 5))

# Plot the number of female students in each class on the first subplot
counts_f.plot(kind='bar', ax=axs[0], color='red', legend=False)
axs[0].set_title('Female students by class')
axs[0].set_xlabel('Class')
axs[0].set_ylabel('Number of students')

# Add annotations to the first subplot
for i, v in enumerate(counts_f):
    axs[0].text(i, v+3, str(v), ha='center')

# Plot the number of male students in each class on the second subplot
counts_m.plot(kind='bar', ax=axs[1], color='blue', legend=False)
axs[1].set_title('Male students by class')
axs[1].set_xlabel('Class')
axs[1].set_ylabel('Number of students')

# Add annotations to the second subplot
for i, v in enumerate(counts_m):
    axs[1].text(i, v+3, str(v), ha='center')
    
# Set the title for the overall figure and adjust the spacing and font
fig.suptitle('Number of students by gender and class', fontsize=16)
fig.subplots_adjust(top=0.85)

# Show the plot
plt.show()
```
